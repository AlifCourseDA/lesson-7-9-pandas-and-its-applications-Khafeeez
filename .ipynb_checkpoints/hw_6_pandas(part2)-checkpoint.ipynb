```python
# Reading csv file with zip compression
df3 = pd.read_csv(r'C:\Users\99899\Downloads\why.zip',encoding='windows-1251', compression = 'zip')
df3
```

```python
df3.dtypes
```

```python
# duplicate finding
# duplicate_num = df3.duplicated().sum()
# duplicate_num
df3.drop_duplicates()



```

```python
# when InvoiceNo contains C means Cancelled so that finding out the Cancelled Transactions
df3.InvoiceNo.str.contains('C').sum()
```

```python
# filtering out with Quantity greater than 0
df3 = df3[df3['Quantity'] > 0]
```

```python
# Filtering Customers from Germany
by_country = df3 \
    .query("Country == 'Germany'") \
    .aggregate({'CustomerID': 'count'}) 
by_country
```

```python
# percentile 80 of filtered users used in unique in germany top
percentile_80 = df3['InvoiceNo'].value_counts().quantile(0.8)
filtered_users = df3['InvoiceNo'].value_counts()[df3['InvoiceNo'].value_counts() > percentile_80]
germany_top = filtered_users.index.unique()
germany_top
```

```python
# Customer ID in with InvoiceNo in germany top column
top_retail_germany = df3[df3['InvoiceNo'].isin(germany_top)]
top_retail_germany
```

```python
# grouping by Stockcode and Description with Quantity
df3_g = df3 \
    .groupby(['StockCode', 'Description'], as_index=False) \
    .aggregate({'Quantity': 'count'}) \
    .sort_values('Quantity', ascending=False)
df3_g
```

```python
# Finding out the Revenue with merging Quantity * UnitPrice
df3.loc[:, 'Revenue'] = df3['Quantity'] * df3['UnitPrice']
df3
```

```python
# Total_Revenue from InvoiceNo
df3_total_Revenue = df3 \
    .groupby('InvoiceNo')['Revenue'].sum() 

df3_total_Revenue = df3_total_Revenue.sort_values(ascending=False)
df3_total_Revenue
```
